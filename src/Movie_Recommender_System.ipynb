```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# def concast(obj):
#     L = []
#     for i in ast.literal_eval(obj):
#         L.append(i['character'])
#     return L[0:3]
def concast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            L.append(i['character'])
            count += 1
        else: 
            break
    return L
```

```python
movies['cast'].apply(concast)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split(" "))
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
final_df = movies[['movie_id', 'title', 'tags']]
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split(" "):
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loving')
```

```python
final_df['tags'] = final_df['tags'].apply(stem)
```

```python
# now we will convert tag text into vector and this technique is known as text vectorisation.
# it can be done by using:
# 1. bag of words
# 2. tf idf
# 3. word2vec
# here we will use bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(final_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarities[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    mov_index = final_df[final_df['title'] == movie].index[0]
    distances = similarities[mov_index]
    mov_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:15]
    for i in mov_list:
        print(final_df.iloc[i[0]].title)
        # print(i[0])
```
